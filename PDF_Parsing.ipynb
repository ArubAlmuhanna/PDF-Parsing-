The below code is written to extract the tables from the PDF file `Recommended Food Storage Times Cold and Dry Refrigerated and Frozen Foods.PDF`. The final output of the code is `Food_Storage.csv`

```python
#import required libraries 
import tabula
from tabula import read_pdf
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
#open the PDF file to extract tables 
pdfFile=tabula.read_pdf("Recommended Food Storage Times Cold and Dry Refrigerated and Frozen Foods .pdf"
                        , pages = "all", multiple_tables = True)
```

```python
len(pdfFile)
```

#### Table 1

```python
dairyTable=pdfFile[0]
dairyTable
```

```python
#fix the ambigous values in the table

#add the Hard cheese the column 'Food Product' 
dairyTable['Food Product'] = np.where((dairyTable['Refrigerator Storage']=='3-4 weeks (opened)'),
                            'Hard cheese',
                            dairyTable['Food Product'])

#add the storage details of the product Hard cheese to the column 'Refrigerator Storage' 
dairyTable['Freezer Storage'] = np.where((dairyTable['Refrigerator Storage']=='3-4 weeks (opened)')
                             &(dairyTable['Food Product'] =='Hard cheese'),
                            '6 months',
                            dairyTable['Freezer Storage'])
```

#### Table 2

```python
#rename the second table as eggTable
eggTable=pdfFile[1]
```

```python
#add the word egg to each row for users future search 
eggTable['Food Product'] = 'Egg ' + eggTable['Food Product'].astype(str)
```

#### Table 3

```python
#rename the table
meatTable=pdfFile[2]
```

```python
meatTable
```

```python
#fix the ambigous values in the table

#add the product Canned ham to the column 'Food Product'
meatTable['Food Product'] = np.where((meatTable['Refrigerator Storage']=='3-5 months (opened)')
                           &(meatTable['Freezer Storage'] =='1-2 months'),
                            'Canned ham',
                            meatTable['Food Product'])

#add the product Hotdogs to the column 'Food Product'
meatTable['Food Product'] = np.where((meatTable['Refrigerator Storage']=='1 week (opened)')
                           &(meatTable['Freezer Storage'] =='1-2 months'),
                            'Hotdogs',
                            meatTable['Food Product'])

#add the product Lunch meats to the column 'Food Product'
meatTable['Food Product'] = np.where((meatTable['Refrigerator Storage']=='3-5 days (opened)')
                           &(meatTable['Freezer Storage'] =='1-2 months'),
                            'Lunch meats',
                            meatTable['Food Product'])

#add the storage details of the product USDA seal to the column 'Refrigerator Storage' 
meatTable['Refrigerator Storage'] = np.where((meatTable['Food Product']=='USDA seal'),
                            '2 weeks (unopened)',
                            meatTable['Refrigerator Storage'])

#add the storage details of the product USDA seal to the column 'Freezer Storage'
meatTable['Freezer Storage'] = np.where((meatTable['Food Product']=='USDA seal'),
                            'Do not freeze',
                            meatTable['Freezer Storage'])

#add the storage details of the product stews to the column 'Refrigerator Storage' 
meatTable['Refrigerator Storage'] = np.where((meatTable['Food Product']=='stews'),
                            '3-4 days',
                            meatTable['Refrigerator Storage'])

#add the storage details of the product stews to the column 'Freezer Storage'
meatTable['Freezer Storage'] = np.where((meatTable['Food Product']=='stews'),
                            '2-3 months',
                            meatTable['Freezer Storage'])
```

#### Table 5

```python
#rename table
fishTable=pdfFile[6]
```

```python
#delete the empty column 
del fishTable['Unnamed: 0'] 
```

```python
fishTable
```

```python
#clean the row that contains Lean Fish in 'Food Product' column 
fishTable['Food Product'] = np.where((fishTable['Refrigerator Storage']=='1-2 days')
                           &(fishTable['Freezer Storage'] =='6 months'),
                            'Lean Fish',
                            fishTable['Food Product'])

#clean the row that contains Fatty fish in 'Food Product' column 
fishTable['Food Product'] = np.where((fishTable['Refrigerator Storage']=='1-2 days')
                           &(fishTable['Freezer Storage'] =='2-3 months'),
                            'Fatty fish',
                            fishTable['Food Product'])

#split the multiple products in the second row by adding them to seperate multiple rows
#salmon
fishTable = fishTable.append({'Food Product': 'salmon', 'Refrigerator Storage' : '1-2 days'
                             ,'Freezer Storage':'2-3 months'} , ignore_index=True)
#tuna
fishTable = fishTable.append({'Food Product': 'tuna', 'Refrigerator Storage' : '1-2 days'
                             ,'Freezer Storage':'2-3 months'} , ignore_index=True)
#swordfish
fishTable = fishTable.append({'Food Product': 'swordfish', 'Refrigerator Storage' : '1-2 days'
                             ,'Freezer Storage':'2-3 months'} , ignore_index=True)

#clean the column 'Refrigerator Storage' and 'Freezer Storage' for the product Clams 
fishTable['Refrigerator Storage'] = np.where((fishTable['Food Product']=='Clams'),
                            '1-2 days',
                            fishTable['Refrigerator Storage'] )

fishTable['Freezer Storage'] = np.where((fishTable['Food Product']=='Clams'),
                            '3-6 months',
                            fishTable['Freezer Storage'] )

#clean the column 'Refrigerator Storage' and 'Freezer Storage' for the product Clams 
fishTable['Refrigerator Storage'] = np.where((fishTable['Food Product']=='Oysters'),
                            '1-2 days',
                            fishTable['Refrigerator Storage'] )

fishTable['Freezer Storage'] = np.where((fishTable['Food Product']=='Oysters'),
                            '3-6 months',
                            fishTable['Freezer Storage'] )

#clean the column 'Refrigerator Storage' and 'Freezer Storage' for the product Mussels 
fishTable['Refrigerator Storage'] = np.where((fishTable['Food Product']=='Mussels'),
                            '1-2 days',
                            fishTable['Refrigerator Storage'] )

fishTable['Freezer Storage'] = np.where((fishTable['Food Product']=='Mussels'),
                            '3-6 months',
                            fishTable['Freezer Storage'] )
```

```python
#split the multiple products in the tenth row by adding them to seperate multiple rows

#honey
MiscellaneousTable['Food Product'] = np.where((MiscellaneousTable['Refrigerator Storage']=='6-8 months (opened)')
                           &(MiscellaneousTable['Freezer Storage'] =='Freeze poorly'),
                            'honey',
                            MiscellaneousTable['Food Product'])
#jams
MiscellaneousTable = MiscellaneousTable.append({'Food Product': 'jams', 'Refrigerator Storage' : '6-8 months'
                             ,'Freezer Storage':'Freeze poorly'} , ignore_index=True)
#jellies
MiscellaneousTable = MiscellaneousTable.append({'Food Product': 'jellies', 'Refrigerator Storage' : '6-8 months'
                             ,'Freezer Storage':'Freeze poorly'} , ignore_index=True)
#syrup
MiscellaneousTable = MiscellaneousTable.append({'Food Product': 'syrup', 'Refrigerator Storage' : '6-8 months'
                             ,'Freezer Storage':'Freeze poorly'} , ignore_index=True)
```

Merge all the tables in one table

```python
#combine all the tables into one list 
pdfFinal=[dairyTable,eggTable,meatTable,PoultryTable,
          fishTable,fruiteTable,VegetablesTable,
          BakedTable,CakesCookiesTable,PastriesPiesTable,
         BabyFoodTable,MiscellaneousTable]
```

```python
#concatenate the lists
pdfFinal = pd.concat(pdfFinal)
```

```python
#save to csv file
pdfFinal.to_csv('Food_Storage.csv',index=False)
```
